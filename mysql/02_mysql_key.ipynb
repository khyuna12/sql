#### buyTbl에 데이터를 추가
- 외래키로 지정되어 있는 userID에 입력되는 새로운 값 STJ가 userTbl에 없는 값이어서 무결성 오류 발생

```python
cursor = db.cursor()
sql_query = "INSERT INTO buyTbl (userID, prodName, groupName, price, amount) VALUES('SSS', '운동화', NULL, 30, 2)"
cursor.execute(sql_query)
db.commit()  # 에러나는게 정상
```

```python
cursor = db.cursor()
SQL_QUERY = "INSERT INTO buyTbl (userID, prodName, groupName, price, amount) VALUES('BBK', '운동화', NULL, 30, 2)"
cursor.execute(SQL_QUERY)
db.commit()
```

```python
db.close()
```

```python
# db 연결을 활성화해주는 함수 구현
def conn(d_name):
    import pymysql
    host_name = 'localhost'
    host_port = 3307
    username = 'root'
    password = 'toor'
    database_name = d_name
    db = pymysql.connect(
        host=host_name,
        port=host_port,
        user=username,
        passwd=password,
        db=database_name,
        charset='utf8'
    )
    return db
```

```python
db = conn('sqlDB')
```

```python
# userTbl에 userID가 STJ인 데이터를 넣어준 후에, 다시 buyTbl userID에 STJ 관련 데이터 넣기
cursor = db.cursor()
sql_query = "INSERT INTO userTbl VALUES('SSS', '서태지', 1975, '경기', '011', '00000000', 171, '2012-4-4')"
cursor.execute(sql_query)
db.commit()
```

```python
SQL_QUERY = "INSERT INTO buyTbl (userID, prodName, groupName, price, amount) VALUES('SSS', '운동화', NULL, 30, 2)"
cursor.execute(SQL_QUERY)
db.commit()
```

```python
# userTbl에 userID가 STJ 관련 데이터 삭제
sql_query = "delete from userTbl where userID = 'SSS'"
cursor.execute(sql_query)
db.commit()  # 에러나는게 정상 - buyTbl에 해당 userID를 참조하는 데이터가 있기 때문
```
